### 正式处理

##### 子公司列表处理

```stata
use $subcom, clear
ren FN_Fn06101 user
exm_com_name user
ren FN_Fn06104 indtype
ren FN_Fn06107 subcom_overseas
ren FN_Fn06117 subcom_exit
* 步骤1：将字符串转为Stata日期格式
gen date_numeric = date(EndDate, "YMD")  // "YMD"表示年月日顺序
format date_numeric %td  // 设置为日期格式
* 步骤2：提取年份
gen year = year(date_numeric)
recast int year, force
* 清理中间变量（可选）
drop date_numeric
duplicates drop user year, force
drop FN_Fn06105 FN_Fn06106 FN_Fn06108
save subcom, replace
```

```stata
use $listcom, clear
keep id id_org  year ///面板标识
FullName        ListedCoID      Stknme ///企业身份
Markettype      Aonly       ///资本市场性质
Estbyear        Listyear    Ipoyear     age     lnage ///年份和年龄
govcon1_p       govcon2_p   hierarchy ///行业分类
PROVINCECODE    CITYCODE    Lng Lat ///地理位置
TA size TD      Sale        lnSale ///财务绝对值
tl sl Finlev ///负债率
tang itang liqui  cr cflow cash invt arnr ///资产率
roa roe //利润率
rename id_org Stkcd
drop if missing(govcon1_p) & missing(govcon2_p)
merge 1:m Stkcd year using subcom
keep if _merge == 3
drop _merge
save com, replace
```

## 指标计算

```stata
use merged-exm, clear
gen landprice = price / realarea
gen buildprice = price / building_area
label var landprice "地面价（公顷/万元）"
label var buildprice "楼面价（公顷/万元）"
gen landprice_yuan = 10000*landprice
gen buildprice_yuan = 10000*buildprice
gen ln_landprice = ln(landprice_yuan)
gen ln_buildprice = ln(buildprice_yuan)
drop landprice_yuan
drop buildprice_yuan
label var ln_landprice "自然对数：地面价（公顷/元）"
label var ln_buildprice "自然对数：楼面价（公顷/元）"
save merged-computed, replace
```

## 地市主官面板

```stata
use $official, clear
gen mayor = 1 if (local_job_title1 == 15 | local_job_title2 == 15 ///
| local_job_title3 == 15 | local_job_title4 == 15 ) //市长
replace mayor = 0 if (local_job_title1 == 14 ///
| local_job_title2 == 14 | local_job_title3 == 14 ) ///
| local_job_title4 == 14 //市委书记
drop if missing(mayor)
label var mayor "主官职务(1=市长|0=书记)"
drop local_job_title*
* 步骤1: 生成年份序列变量并扩展数据集
gen len = end_yr - begin_yr + 1  // 计算每个任期覆盖的年数
expand len                       // 根据年数扩展观测值
label var len "任期年数"

* 步骤2: 为每个扩展观测生成对应的年份
bysort officerID begin_yr end_yr: gen year = begin_yr + _n - 1

* 步骤3: 处理同年同地区同职务的重叠
   // 排序规则：优先保留begin_yr更小的官员（资历更老），其次end_yr更大的官员，最后按officerID任意排序
gsort region_code mayor year +begin_yr -end_yr +officerID

   // 标记并保留每组（region_code-mayor-year）的第一个观测值（即最早开始任职者）
bysort region_code mayor year: gen keep_flag = (_n == 1)
keep if keep_flag
drop keep_flag

* 步骤4: 最终排序并整理变量
order region_code mayor year officerID begin_yr end_yr
sort region_code mayor year
```
